```python
productsFilePath = "abfss://miniproject@unluckystorageaccount2.dfs.core.windows.net/products.csv"
display(productsFilePath)

ordersFilePath = "abfss://miniproject@unluckystorageaccount2.dfs.core.windows.net/orders.csv"
display(ordersFilePath)

customersFilePath = "abfss://miniproject@unluckystorageaccount2.dfs.core.windows.net/customers.csv"
display(customersFilePath)
```

```python
productsDF = spark.read.format("csv").option("header",True).option("inferSchema",True).load(productsFilePath)
#display(productsDF)
ordersDF = spark.read.format("csv").option("header",True).option("inferSchema",True).load(ordersFilePath)
#display(ordersDF)
customersDF = spark.read.format("csv").option("header",True).option("inferSchema",True).load(customersFilePath)
#display(customersDF)
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import *
```

```python
#1.	Create a column PNAME_UPPER that converts PNAME to uppercase using upper().
productsDF = productsDF.withColumn("PNAME_UPPER",upper("PNAME"))
display(productsDF)
```

```python
#2.	Create a column PROD_CODE that extracts the first 4 characters of PRODID using substring().
productsDF = productsDF.withColumn("PROD_CODE",substring("PRODID",1,4))
display(productsDF)
```

```python
productsDF.printSchema()
```

```python
#3.	Create a column PRODUCT_INFO that combines PNAME and CATEGORY separated by ' - ' using concat_ws().
productsDF = productsDF.withColumn("PRODUCT_INFO", concat_ws(" - ", col("PNAME"), col("CATEGORY")))
display(productsDF)
```

```python
#3.	Create a column PRODUCT_INFO that combines PNAME and CATEGORY separated by ' - ' using concat_ws().
productsDF = productsDF.withColumn("PRODUCT_INFO", concat_ws("-", col("PNAME"), col("CATEGORY")))
display(productsDF)
```

```python
#4.	Convert ORDERDATE (format dd-MMM-yy) in orders to a proper date type using to_date().
ordersDF = ordersDF.withColumn("ORDERDATE",to_date("ORDERDATE","dd-MMM-yy"))
display(ordersDF)
```

```python
#5.	Using the converted ORDERDATE, create a column DAYS_SINCE_ORDER showing the number of days between ORDERDATE and current_date().
ordersDF = ordersDF.withColumn("DAYS_SINCE_ORDER", datediff(current_date(), col("ORDERDATE")))
display(ordersDF)
```

```python
#6.	Replace null values in DISCOUNT with 0 using fillna().
ordersDF = ordersDF.fillna({'DISCOUNT': 0})
display(ordersDF)
```

```python

#7.	Create a column DISCOUNT_SAFE using coalesce() that returns DISCOUNT if present, otherwise 0.
colordersDF = ordersDF.withColumn("DISCOUNT_SAFE", coalesce(col("DISCOUNT"), lit(0)))

display(ordersDF)
```

```python

#8.	On orders, create a column ORDER_TYPE using when()/otherwise() that maps STATUS 'DELIVERED'->'Success', 'CANCELLED'->'Failed', 'RETURNED'->'Reversed', and anything else->'Unknown'.
ordersDF = (ordersDF.withColumn("ORDER_TYPE",
                               when(col("STATUS")=='DELIVERED' , "Sucess")
                               .when(col("STATUS")=="CANCELLED","Failed")
                               .when(col("STATUS")=='RETURNED',"Reversed")
                               .otherwise("Unknown"))
            )
display(ordersDF)            

```
