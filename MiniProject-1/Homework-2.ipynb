```python
productsFilePath = "abfss://miniproject@unluckystorageaccount2.dfs.core.windows.net/products.csv"
display(productsFilePath)

ordersFilePath = "abfss://miniproject@unluckystorageaccount2.dfs.core.windows.net/orders.csv"
display(ordersFilePath)

customersFilePath = "abfss://miniproject@unluckystorageaccount2.dfs.core.windows.net/customers.csv"
display(customersFilePath)
```

```python
productsDF = spark.read.format("csv").option("header",True).option("inferSchema",True).load(productsFilePath)
#display(productsDF)
ordersDF = spark.read.format("csv").option("header",True).option("inferSchema",True).load(ordersFilePath)
#display(ordersDF)
customersDF = spark.read.format("csv").option("header",True).option("inferSchema",True).load(customersFilePath)
#display(customersDF)
```

```python
customersDF.printSchema()
productsDF.printSchema()
ordersDF.printSchema()
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import *
```

```python
#1.	Filter products whose PRICE is greater than 1000.
productsDF =productsDF.filter(col("PRICE") > 1000) 
display(productsDF)
```

```python
#2.	Filter orders where QTY is greater than 5 AND STATUS is 'DELIVERED' (use & operator).
ordersDF = ordersDF.filter((col("QTY") > 5)  & (col("STATUS") == "DELIVERED"))
display(ordersDF)
```

```python
#3.	Filter products whose CATEGORY is either 'Electronics' or 'Furniture' using isin().
productsDF = productsDF.filter(col("CATEGORY").isin("Electronics","Furniture"))
display(productsDF)
                       
```

```python
#4.	Filter orders where DISCOUNT is null (no discount applied).
ordersDF = ordersDF.filter(col("DISCOUNT").isNull())
display(ordersDF)
```

```python
#5.	Find the distinct CATEGORY values present in products.
productsDF_dist= productsDF.select("CATEGORY").distinct()
display(productsDF_dist)
```

```python
#6.	Remove duplicate rows from orders based on CUSTID (keep only one order per customer).
customersDF = customersDF.dropDuplicates(["CUSTID"])
display(customersDF)
```

```python
#7.	Remove duplicate rows from orders based on CUSTID (keep only one order per customer).
productsDF = productsDF.orderBy("PRICE",ascending=False)
display(productsDF)
```

```python
#8.	Sort orders first by STATUS (ascending) and then by QTY (descending).
ordersDF =ordersDF.orderBy("STATUS",ascending=True).orderBy("QTY",ascending=False)
display(ordersDF)
```

```python
#2nd Method
ordersDF = ordersDF.orderBy(["STATUS", "QTY"], ascending=[True, False])
display(ordersDF)
```
